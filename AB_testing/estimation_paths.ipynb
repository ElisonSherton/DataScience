# WIP: Estimation Paths

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_samples = 500
for _ in range(20):
    x = np.random.binomial(1, 0.5, n_samples)
    difference = np.array([])
    for i, val in enumerate(x):

        prob_head = np.sum(x[:i + 1]) / len(x[:i + 1])
        prob_differnce = prob_head - (1 - prob_head)
        difference = np.append(difference, [prob_differnce]) 
    plt.plot(np.arange(n_samples), difference, 'k', alpha=0.3, color='#6596ff')
        
plt.ylim([-1, 1])
plt.xlim([0, n_samples])
plt.show()
```
